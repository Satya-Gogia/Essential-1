## INPUT FUNCTION

- The input function is used to read data from user and return the data to the running program.
- The program can manipulate the data and make code truly intervactive.
- A program that does not get a user's input is a def program

```python
print ("Tell me anything..")
data=input()
print("Hmm.. ", data," \n ..really")
```

 ## INPUT FUNCTION WITH AN ARGUMENT

 - The input function also be used with an argument and not with the help of a print function 
 - The argument given in the input function is a string and gets displayed to the console before user has the chance to input anything
 - This completely eliminates the use of the print function being associated to the input function

```python
anything = input("Tell me anything...")
print("Hmm...", anything, "...Really?")

```

## NOTE
- The result of a input function is by default a string and thus cannot be used in numeric or float data manipulation as is 
- For using input function to help in numeric operations we use type casting 
- The code below will throw TypeError message as the data being input is a string but is being used in an arithmetic or numeric operations

```python
anything = input("Enter a number: ")
something = anything ** 2.0
print(anything, "to the power of 2 is", something)
```

## TYPE CASTING (TYPE CONVERSION)
- There are two simple functions that python offers 
 - int()
 - float()
- This functions takes string as an argument and converts it into the type it is being specified into 

```python
anything = float(input("Enter a number: "))
something = anything ** 2.0
print(anything, "to the power of 2 is", something)
```

```python
# Length of hypotenuse
leg_a = float(input("input first leg length : "))
leg_b = float(input("input second leg length : "))
hyp = (leg_a ** 2 + leg_b ** 2) ** .5
print ( "hypotenuse length is : ", hyp)
```

## STRING OPERATIONS
- Conecatination operator (+) is used for joining to strings together 
- It simply glues 2 strings together 
- It is not commutative so ("ab" + "ba") is not equal to ("ba" + "ab")
- The rule for concatination is that both vaules should be string

```python
fnam = input("May I have your first name, please? ")
lnam = input("May I have your last name, please? ")
print("Thank you.")
print("\nYour name is " + fnam + " " + lnam + ".")
```

- Replication operator (*) is used for repeating the string as many times as desired
- It is commutative as ("stirng" * number) is equal to (number * "string")
- The rule for concationation is that one should be a string and other should be a number or it would not suffice

```python
print("+" + 10 * "-" + "+")
print(("|" + " " * 10 + "|\n") * 5, end="")
print("+" + 10 * "-" + "+")
```
